```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
SP500_file = r"C:\Users\KirkLaptop\Documents\Bootcamp\Resources\all_stocks_5yr.csv"
SP500_df = pd.read_csv(SP500_file)
SP500_df.head()
```

```python
stockname_file = r"C:\Users\KirkLaptop\Documents\Bootcamp\Resources\symbols_valid_meta.csv"
stockname_df = pd.read_csv(stockname_file)
stockname_df.head()
```

```python
new_SP500_df = SP500_df[['date', 'open', 'close', 'volume', 'Name']].copy()
new_SP500_df.head()
```

```python
new_SP500_df.rename(columns = {'Name':'Symbol'}, inplace = True)
new_SP500_df.head()
```

```python
new_stockname_df = stockname_df[['Symbol', 'Security Name', 'Listing Exchange']].copy()
new_stockname_df.head()
```

```python
connection_string = "postgres:SQL4062!@localhost:5432/Stonks"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
engine.table_names()
```

```python
new_SP500_df.to_sql(name='SP500', con=engine, if_exists='append', index=False)
```

```python
new_stockname_df.to_sql(name='stockname', con=engine, if_exists='append', index=False)
```
